**Mat Data using Python**

Mat files: Files with the .mat extension are files that are in the binary data container format that the MATLAB program uses. The extension was developed by Mathworks and MAT files are categorized as data files that include variables, functions, arrays and other information. MAT files can be saved in a variety of formats as long as users choose to save them in a version that belongs to the MATLAB Preferences. There are the so called level 4 MAT files wherein two-dimensional matrices and character strings are supported and there is also the level 5 MAT files wherein several things are included like the cell arrays, objects, multidimensional numeric arrays, structures and character arrays. MAT files are also useful when it comes to representing audio in various formats such as 16-bit signed integer, 8-bit unsigned integer and 64-bit floating point. Mathworks MATLAB is the software used to open MAT files. It is an application used to develop algorithm, visualize and analyze data as well as to compute numbers. 


In order to access mat files through python we can use scipy

```python
import scipy.io as scio
import os,pandas as pd
path_to_datasets=os.path.join('..','datasets','umontreal')

directories=os.listdir(path_to_datasets)
current_subject=directories[0]
current_path=os.path.join(path_to_datasets,current_subject)

formatted_data={}

for subfile in os.listdir(current_path):
    if subfile.removesuffix('.mat') not in formatted_data:
        formatted_data[subfile.removesuffix('.mat')]=list()
    
    formatted_data[subfile.removesuffix('.mat')]=list(scio.loadmat(os.path.join(current_path,subfile))['seq_cloud'])

    print(f'{subfile=}')


csv_data={}
for subfile,data in formatted_data.items():
    print(data)
    for x in data:
        print(type(x))
    csv_data[subfile+'X']=[x[0] for x in data]
    csv_data[subfile+'Y']=[x[1] for x in data]
    csv_data[subfile+'Z']=[x[2] for x in data]

print(list(csv_data.keys()))
print(f'Columns:{len(formatted_data)}')

print(pd.DataFrame.from_dict(formatted_data))
```
