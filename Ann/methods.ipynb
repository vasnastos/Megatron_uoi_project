```python
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer

import pandas as pd,os
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import numpy as np
from scipy import stats



# 1. Load dataset
data=pd.read_csv(os.path.join('..','datasets','forceplate.csv'))
print("Dataset description")
print('*'*20)
print(data.describe())
print(f'{data.shape=}')
print('-'*5+' dataset '+'-'*5)
# 2. Normalize values
normalize='MinMax'
# normalize='Z-score'

if normalize=='MinMax':
    scaler=MinMaxScaler()
elif normalize=='Z-score':
    scaler=StandardScaler()
data=pd.DataFrame(data=scaler.fit_transform(data.to_numpy()),columns=data.columns.to_list())

# 3. Data distribution
fig=plt.figure(figsize=(14,10))
grid=gridspec.GridSpec(4,4)
cntIndex=0
cntColumn=0
rowLimit=4
for index,column in enumerate(data.columns.to_list()):
    if index<2:
        continue
    
    if cntColumn%rowLimit==0 and cntColumn!=0:
        cntColumn=0
        cntIndex+=1

    column_data=data[column].to_list()
    ax=fig.add_subplot(grid[cntIndex,cntColumn])
    ax.set_title(f'Line plot for {column}',c='darkgreen',fontdict={"weight":"bold","size":9})
    ax.plot(np.arange(1,len(column_data)+1),column_data,f'{column}',linewidth=2,c='darkblue')
    ax.set_xlabel("Sample",fontdict={'weight':'bold','size':11},labelpad=10)
    ax.set_ylabel(f'{column}',fontdict={'weight':'bold','size':11},labelpad=10)
    ax.set_yticks(np.arange(0,max(data[column].to_list())+20,0.5))


# 4. Statistic analysis

# Z Score
# Z test
# T Test
# F Test
# Correlation matrix







```
